````python
import datetime

# Create the YouTube API service object
youtube = build('youtube', 'v3', developerKey=api_key)

# Date range
now = datetime.datetime.now()
one_week_ago = now - datetime.timedelta(days=7)

# Format the dates according to the API's format
now_str = now.strftime("%Y-%m-%dT%H:%M:%SZ")
one_week_ago_str = one_week_ago.strftime("%Y-%m-%dT%H:%M:%SZ")

# Make the request to retrieve the most popular videos from the past week
response = youtube.search().list(
    part="snippet",
    maxResults=10,  # Get the latest 10 data
    order="viewCount",
    publishedAfter=one_week_ago_str,
    publishedBefore=now_str
).execute()

cont = 0
print("\t\t------------------ TITLES ------------------ \n")
for item in response['items']:
    video_id = item['id']['videoId']
    cont += 1
    # Get video details using the YouTube API
    video_response = youtube.videos().list(
        part='snippet',
        id=video_id
    ).execute()

    # Get the video title and description
    video_title = video_response['items'][0]['snippet']['title']
    video_id = item['id']['videoId']

    # Print the video title and description
    print(cont, ".- {}".format(video_title))
    print("\tid: {}".format(video_id))
````

# Get transcript of the videos

---


```
If the video has transcription if it works
```

````python
from youtube_transcript_api import YouTubeTranscriptApi, TranscriptsDisabled, NoTranscriptFound
from googletrans import Translator

cont = 0
print("------------------ TITLES ------------------ ")
for item in response['items']:
    video_id = item['id']['videoId']
    cont += 1
    # Get video details using the YouTube API
    video_response = youtube.videos().list(
        part='snippet',
        id=video_id
    ).execute()

    # Get the video title and description
    video_title = video_response['items'][0]['snippet']['title']
    video_id = item['id']['videoId']

    # Print the video title and description
    print(cont, ".- {}".format(video_title))
    print("id: {}".format(video_id))

    try:
        # Get available transcripts for the video
        transcript_list = YouTubeTranscriptApi.list_transcripts(video_id)

        for transcript in transcript_list:
            lang_code = transcript.language

            if lang_code == 'en':
                # Skip English transcript as we only want to translate from other languages
                continue

            # Get the official language transcript
            transcript_text = transcript.translate('en').fetch()

            print(f"\nTranscription in {lang_code}:")
            for text in transcript_text:
                print(text['text'])

    except (TranscriptsDisabled, NoTranscriptFound):
        print("No transcript found for this video.")

    print('\n')
````

````python
from translate import Translator
from langdetect import detect

class MyTranslator:
    def __init__(self):
        self.text = ''
        # destiny language
        self.to_lang = 'en'
        self.from_lang = ''
        self.translator = None

    def detect_language(self, text):
        detected_lang = detect(text)
        self.from_lang = detected_lang

    def translate(self, text):
        print(self.to_lang, self.from_lang)
        self.translator = Translator(to_lang=self.to_lang, from_lang=self.from_lang)
        self.text = text
        translation = self.translator.translate(text)
        return translation
````

````python
from summa import summarizer

cont = 0
print("------------------ TITLES ------------------ ")
for item in response['items']:
    video_id = item['id']['videoId']
    cont += 1
    # Get video details using the YouTube API
    video_response = youtube.videos().list(
        part='snippet',
        id=video_id
    ).execute()

    # Get the video title and description
    video_title = video_response['items'][0]['snippet']['title']
    video_id = item['id']['videoId']

    # Print the video title and description
    print(cont, ".- {}".format(video_title))
    print("id: {}".format(video_id))

    try:
        # Get available transcripts for the video
        transcript_list = YouTubeTranscriptApi.list_transcripts(video_id)

        for transcript in transcript_list:
            lang_code = transcript.language

            if lang_code == 'en':
                # Skip English transcript as we only want to translate from other languages
                continue

            # Get the official language transcript
            transcript_text = transcript.translate('en').fetch()

            full_transcript_text = "\n".join(text['text'] for text in transcript_text)
            translator = MyTranslator()
            translator.detect_language(full_transcript_text)

            transcript_text = translator.translate(full_transcript_text)

            if not transcript_text:
                print("No transcript available for this video.")
                continue

            '''print(f"\nTranscription in {lang_code}:")
            for text in transcript_text:
                print(text['text'])'''

            # Get a brief summary of the transcript
            #summarized_text = summarizer.summarize("\n".join(text['text'] for text in transcript_text))
            summarized_text = summarizer.summarize(full_transcript_text)
            print("\nTranscription Summary:")
            print(summarized_text)

    except (TranscriptsDisabled, NoTranscriptFound):
        print("No transcript found for this video.")

    print('\n')
````
